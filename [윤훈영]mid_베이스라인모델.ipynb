```python
import re
def remove(col):
    pattern = r'\([^)]*\)'
    col = [re.sub(pattern, '', item).strip() for item in col if item and not re.search(pattern, item)]
    return col
```

```python
df['중식메뉴'] = df['중식메뉴'].apply(remove)
df['석식메뉴'] = df['석식메뉴'].apply(remove)
df.head(3)
```

```python
#메뉴 분리 및 메인메뉴만 수합
#민규님 코드 참고
def menu_split(df, col):
    rice = [row[0] if len(row) > 2 else row for row in df[col].values]
    soup = [row[1] if len(row) > 2 else row for row in df[col].values]
    main = [row[2] if len(row) > 2 else row for row in df[col].values]
    df[f'{col[:2]}_밥'] = rice
    df[f'{col[:2]}_국'] = soup
    df[f'{col[:2]}_메인'] = main
    return df
```

```python
def menu_processing(df, col):
    rice = []
    for row in df[col].values:
        if '/' in row:
            rice.append(row.split('/')[0].strip())
        elif '*' in row:
            try: 
                rice.append(row.split('*')[0].strip())
            except:
                rice.append(str(row).split('*')[0].strip())
        else:
            try:
                rice.append(row.strip())
            except:
                rice.append(row)
    df[col] = rice
    return df
```

```python
#밥, 국은 큰 의미가 없을 것 같으므로 중식_메인, 석식_메인값만 인코딩을 진행해준다
#각 메뉴별 타깃값의 평균값을 구해서 10등분을 하고, 상위 1위부터 1~10까지 순위를 부여하자
import numpy as np #time은 석식
def rank(df, time):
    df_temp = df.copy()
    df_temp = df_temp.dropna(subset=f'{time}계')
    df_temp = df_temp[df_temp[f'{time}계']!=0 | df_temp[f'{time}계'].isna()]
    df_temp2 = df_temp[['일자', f'{time}계', f'{time}_메인']]
    temp = df_temp2.groupby([f'{time}_메인'])[f'{time}계'].mean().reset_index()
    temp = temp.sort_values(by=f'{time}계', ascending=False)
    temp[f'{time}_순위'] = pd.qcut(temp[f'{time}계'], q=10, labels=range(1, 11))
    temp[f'{time}_순위'] = temp[f'{time}_순위'].astype(int)
    df_merged = pd.merge(df_temp, temp[[f'{time}_메인', f'{time}_순위']], on=f'{time}_메인', how='left')
    return df_merged
```

```python
석식_순위 = rank(df, '석식')[['일자','석식_순위']]
중식_순위 = rank(df, '중식')[['일자','중식_순위']]
df = pd.merge(df, 중식_순위, on='일자', how='left')
df2 = pd.merge(df, 석식_순위, on='일자', how='left')
df2.head(3)
```
